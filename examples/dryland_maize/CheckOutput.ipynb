```python
#######
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df_2001_c = pd.read_csv("./outputs/02001sccd1", delim_whitespace=True, header=0)
df_2002_c = pd.read_csv("./outputs/02002sccd1", delim_whitespace=True, header=0)
df_2003_c = pd.read_csv("./outputs/02003sccd1", delim_whitespace=True, header=0)

fig = plt.figure(figsize=(6.5, 7.5))  # Adjust the figure size as needed
ax1 = fig.add_subplot(311)
var='CO2_FLUX'
ax1.plot(np.arange(len(df_2001_c[var])), df_2001_c[var],color='red', label='2001')
ax1.plot(np.arange(len(df_2002_c[var])), df_2002_c[var],color='blue', label='2002')
ax1.plot(np.arange(len(df_2003_c[var])), df_2003_c[var],color='black', label='2003')
ax2 = fig.add_subplot(312)
var='NBP'
ax2.plot(np.arange(len(df_2001_c[var])), df_2001_c[var],color='red', label='2001')
ax2.plot(np.arange(len(df_2002_c[var])), df_2002_c[var],color='blue', label='2002')
ax2.plot(np.arange(len(df_2003_c[var])), df_2003_c[var],color='black', label='2003')
```
